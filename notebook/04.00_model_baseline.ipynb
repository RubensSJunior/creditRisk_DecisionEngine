# Models

```python
with mlflow.start_run(run_name="baseline_logistic_regression"):
    baselineModel = LogisticRegression(
    max_iter=1000,
    class_weight='balanced',
    verbose=0  )

    baselineModel.fit(X_train, y_train)
    y_pred_proba = baselineModel.predict_proba(X_test)[:, 1]
    auc = roc_auc_score(y_test, y_pred_proba)
    
    # 🔹 Log params
    mlflow.log_param("model", "logistic_regression")
    mlflow.log_param("class_weight", "balanced")
    mlflow.log_param("max_iter", 1000)
    
    # 🔹 Log metric
    mlflow.log_metric("auc", auc)
    mlflow.sklearn.log_model(baselineModel, name="LogisticRegressionModel", input_example=X_test)
```

## BaseLine LR

```python
import os
from pathlib import Path
import fnmatch


def load_gitignore(root_path):
    gitignore_path = os.path.join(root_path, ".gitignore")
    patterns = []

    if os.path.exists(gitignore_path):
        with open(gitignore_path, "r") as f:
            for line in f:
                line = line.strip()
                if line and not line.startswith("#"):
                    patterns.append(line)

    return patterns


def should_ignore(path, patterns, root_path):
    rel_path = os.path.relpath(path, root_path)

    for pattern in patterns:
        # match direto
        if fnmatch.fnmatch(rel_path, pattern):
            return True
        # match por nome (ex: __pycache__/)
        if fnmatch.fnmatch(os.path.basename(path), pattern):
            return True

    return False


def print_tree(root_path, patterns, prefix=""):
    items = sorted(os.listdir(root_path))
    items = [
        item for item in items
        if not should_ignore(os.path.join(root_path, item), patterns, root_path)
    ]

    pointers = ["├── "] * (len(items) - 1) + ["└── "]

    for pointer, item in zip(pointers, items):
        path = os.path.join(root_path, item)
        print(prefix + pointer + item)

        if os.path.isdir(path):
            extension = "│   " if pointer == "├── " else "    "
            print_tree(path, patterns, prefix + extension)


def generate_project_structure(path):
    patterns = load_gitignore(path)
    print(".")
    print_tree(path, patterns)

```
