## Definição do Problema de Negócio
O objetivo deste projeto é prever o risco de inadimplência de clientes, identificando a probabilidade de um indivíduo apresentar atraso significativo no pagamento de suas obrigações financeiras. Esse tipo de problema é comum em cenários de **credit scoring**, onde instituições financeiras precisam tomar decisões mais seguras na concessão de crédito.

## Fontes de Dados
Os dados utilizados neste projeto são provenientes do dataset **Give Me Some Credit**, amplamente utilizado para problemas de risco de crédito.

- Tipo de fonte: arquivo estruturado (CSV)
- Origem: base pública para modelagem de crédito
- Conteúdo: informações financeiras e comportamentais dos clientes

## Extração dos Dados
